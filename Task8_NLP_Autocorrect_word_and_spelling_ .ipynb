```python
!pip install pyenchant
import enchant

# Initialize the dictionary
dictionary = enchant.Dict("en_US")

# Define a function to correct the spelling of a word
def correct_word(word):
    if dictionary.check(word):
        return word
    else:
        suggestions = dictionary.suggest(word)
        if len(suggestions) > 0:
            return suggestions[0]
        else:
            return word

# Define a function to correct the spelling of a sentence
def correct_sentence(sentence):
    corrected_sentence = []
    words = sentence.split()
    for word in words:
        corrected_word = correct_word(word)
        corrected_sentence.append(corrected_word)
    return ' '.join(corrected_sentence)

# Test the function with 100 random data points
import random
import string

for i in range(100):
    # Generate a random word or sentence
    length = random.randint(1, 10)
    if random.randint(0, 1):
        # Generate a random word
        word = ''.join(random.choices(string.ascii_lowercase, k=length))
        print(f"{word} -> {correct_word(word)}")
    else:
        # Generate a random sentence
        sentence = ' '.join([''.join(random.choices(string.ascii_lowercase, k=random.randint(1, 10))) for j in range(length)])
        print(f"{sentence} -> {correct_sentence(sentence)}")
```
